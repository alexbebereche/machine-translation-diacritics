```python
import os
import pandas as pd
from ludwig.api import LudwigModel
import yaml

data = {
    "phrase": [],
    "diacritics_phrase": []
}

def create_training_set():
    for file_name in os.listdir("./training"):
        print(file_name)
        if file_name.endswith("gt.txt"):
            with open("./training/" + file_name, "r", encoding="utf-8") as f:
                text = f.readline()
                data["diacritics_phrase"].append(text)
        if file_name.endswith("gt.txt.bak"):
            with open("./training/" + file_name, "r", encoding="utf-8") as f:
                text = f.readline()
                data["phrase"].append(text)
            
    df = pd.DataFrame.from_dict(data)
    df.to_csv("data.csv", encoding="utf_8", index=None)
```

```python
create_training_set()
```

```python
model_definition = None
with open("config.yaml", "r") as stream:
    try:
        model_definition = yaml.safe_load(stream)        
    except yaml.YAMLError as exc:
        print(exc)
        
        
model = LudwigModel(model_definition)
data = pd.read_csv("data.csv")

print(data.head(200))

train_stats = model.train(dataset=data)


print(train_stats)
```
